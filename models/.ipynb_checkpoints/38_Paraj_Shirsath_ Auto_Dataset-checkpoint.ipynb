```python
# we fill the missing values 
data['bore'] = data['bore'].fillna(data['bore'].median()).astype(float)
data['stroke'] = data['stroke'].fillna(data['stroke'].median()).astype(float)
data['horsepower'] = data['horsepower'].fillna(data['horsepower'].median()).astype(float)
data['peak-rpm'] = data['peak-rpm'].fillna(data['peak-rpm'].median()).astype(float)
data['price'] = data['price'].fillna(data['price'].median()).astype(float)
```

```python
x=data.drop("price",axis=1)
x.head()
```

```python
y=data["price"]
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Model Training

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# for testing datasets
y_pred=model.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("MSE :",mse)
print("RMSE :",rmse)
print("MAE :",mae)
print("R_squared :",r2)
```

```python
# for training datasets
y_train_pred=model.predict(x_train)
mse=mean_squared_error(y_train,y_train_pred)
rmse=np.sqrt(mse)
mae=mean_absolute_error(y_train,y_train_pred)
r2=r2_score(y_train,y_train_pred)
print("MSE :",mse)
print("RMSE :",rmse)
print("MAE :",mae)
print("R_squared :",r2)
```
